```python
from keras import layers, models
from tensorflow.keras.layers import SimpleRNN, Input, Dense, concatenate, LSTM
from tensorflow.keras.models import Sequential, Model

# Theta
Theta_input = Input(shape=[None, 1225])
Theta_output =Dense(256, activation='relu')(Theta_input)
Theta_output =Dense(256, activation='relu')(Theta_output)
Theta_output =Dense(256, activation='relu')(Theta_output)
Theta_model = Model(inputs = Theta_input, outputs = Theta_output)


# Alpha
Alpha_input = Input([None, 1225])
Alpha_output = Dense(256, activation='relu')(Alpha_input)
Alpha_output = Dense(256, activation='relu')(Alpha_output)
Alpha_output = Dense(256, activation='relu')(Alpha_output)
Alpha_model = Model(inputs = Alpha_input, outputs = Alpha_output)


# Low_beta
Low_beta_input = Input([None, 1225])
Low_beta_output = Dense(256, activation='relu')(Low_beta_input)
Low_beta_output = Dense(256, activation='relu')(Low_beta_output)
Low_beta_output = Dense(256, activation='relu')(Low_beta_output)
Low_beta_model = Model(inputs = Low_beta_input, outputs = Low_beta_output)


# High_beta
High_beta_input = Input([None, 1225])
High_beta_output = Dense(256, activation='relu')(High_beta_input)
High_beta_output = Dense(256, activation='relu')(High_beta_output)
High_beta_output = Dense(256, activation='relu')(High_beta_output)
High_beta_model = Model(inputs = High_beta_input, outputs = High_beta_output)


# Gamma
Gamma_input = Input([None, 1225])
Gamma_output = Dense(256, activation='relu')(Gamma_input)
Gamma_output = Dense(256, activation='relu')(Gamma_output)
Gamma_output = Dense(256, activation='relu')(Gamma_output)
Gamma_model = Model(inputs = Gamma_input, outputs = Gamma_output)





concatenated = concatenate([Theta_model.output, Alpha_model.output, 
                           Low_beta_model.output, High_beta_model.output, 
                           Gamma_model.output])
concatenated = Dense(256, activation='relu')(concatenated)
concatenated = Dense(256, activation='relu')(concatenated)
concat_out = Dense(1, activation='sigmoid')(concatenated)

concat_model = Model([Theta_input, Alpha_input, 
                     Low_beta_input, High_beta_input, 
                    Gamma_input], concat_out)
concat_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

history = concat_model.fit([train_Theta, train_Alpha, 
                           train_Low_beta, train_High_beta, 
                            train_Gamma], train_y, epochs=1000, batch_size=32)
```
